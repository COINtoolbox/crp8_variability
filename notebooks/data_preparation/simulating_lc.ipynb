```python
import glob

# Get all CSV files in the folder
csv_files = glob.glob(os.path.join(folder, "*.csv"))

dfs = []
empty_files = []

for file in csv_files:
    df_tmp = pd.read_csv(file)
    if df_tmp.shape[0] == 0:
        print(f"No rows in file: {os.path.basename(file)}")
    else:
        dfs.append(df_tmp)

# Concatenate all non-empty DataFrames
if dfs:
    df_all = pd.concat(dfs, ignore_index=True)
    display(df_all)
else:
    print("No non-empty files found.")
```

```python
import glob

csv_files = glob.glob(os.path.join(folder, "*.csv"))

for file in csv_files:
    df_tmp = pd.read_csv(file)
    if df_tmp.shape[0] == 0:
        print(f"No rows in file: {os.path.basename(file)}")
        continue
    required_cols = {'mjd', 'mag', 'catflags', 'filtercode'}
    if required_cols.issubset(df_tmp.columns):
        df_good = df_tmp[df_tmp['catflags'] == 0]
        if df_good.shape[0] == 0:
            print(f"No good rows in file: {os.path.basename(file)} (all catflags != 0)")
            continue
        for fcode, group in df_good.groupby('filtercode'):
            plt.figure()
            plt.scatter(group['mjd'], group['mag'], s=10)
            plt.xlabel('MJD')
            plt.ylabel('mag')
            plt.title(f"{os.path.basename(file)} - filtercode: {fcode}")
            plt.show()
    else:
        print(f"File {os.path.basename(file)} missing required columns: {required_cols - set(df_tmp.columns)}")
```
